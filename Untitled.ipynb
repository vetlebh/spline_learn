```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def b(d, t, i, x):
    if(d==0 and (t[i]<x<=t[i+1])):
        return 1
    elif (d==0):
        return 0
    else:
        if (t[i+d]-t[i]==0):
            a = 0
        else:
            a = (x-t[i])/(t[i+d]-t[i])
        if (t[i+d+1]-t[i+1]==0):
            c = 0
        else:
            c = (t[i+d+1]-x)/(t[i+d+1]-t[i+1])
        return a*b(d-1, t, i, x) + c*b(d-1, t, i+1, x)
    
```

```python
t = np.array([0.0, 0.25, 0.5, 0.75, 1.0])
x = np.linspace(0, 1, 100)
y1 = []
y2 = []
y3 = []
for delx in x:
    y1.append(b(1, t, 0, delx))
    y2.append(b(1, t, 1, delx))
    y3.append(b(1, t, 2, delx))
plt.figure()
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.show()
```
