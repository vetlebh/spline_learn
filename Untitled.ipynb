```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def b(d, t, i, x):
    if (d>0):
        if (t[i+d]-t[i]==0):
            a = 0
        else:
            a = (x-t[i])/(t[i+d]-t[i])
        if (t[i+d+1]-t[i+1]==0):
            c = 0
        else:
            c = (t[i+d+1]-x)/(t[i+d+1]-t[i+1])
        return a*b(d-1, t, i, x) + c*b(d-1, t, i+1, x)
    elif (d==0):
        if(t[i]<x<=t[i+1]):
            return 1
        elif (x>t[i+1]):
            return 0
        elif (x<=t[i]):
            return 0
    
```

```python
 t = np.array([0.,         0.08333333, 0.16666667, 0.25,       0.33333333, 0.41666667,
 0.5,        0.58333333, 0.66666667, 0.75,       0.83333333, 0.91666667,
 1.        ])
x = np.linspace(0, 1, 1000)
y1 = []
y2 = []
y3 = []
y4 = []
y5 = []
y6 = []
y7 = []
y8 = []
y9 = []
y10 = []

d = 3

for delx in x:
    y1.append(b(d, t, 0, delx))
    y2.append(b(d, t, 1, delx))
    y3.append(b(d, t, 2, delx))
    y4.append(b(d, t, 3, delx))
    y5.append(b(d, t, 4, delx))
    y6.append(b(d, t, 5, delx))
    y7.append(b(d, t, 6, delx))
    y8.append(b(d, t, 7, delx))
    y9.append(b(d, t, 8, delx))
    #y10.append(b(d, t, 9, delx))
plt.figure()
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.plot(x, y4)
plt.plot(x, y5)
plt.plot(x, y6)
plt.plot(x, y7)
plt.plot(x, y8)
plt.plot(x, y9)
plt.savefig('closed_b.pdf')
```

```python
 def bspline_plot(d):
    c = np.array([10, 0, 7, 3, 1, 4])
    n = c.shape[0]
    t = np.arange(n+d+1)

    def bspline(c, t, x, d):
        n = c.shape[0]
        y = []
        print(f'Nr control points: {n}')
        for delx in x:
            y_k = 0
            for i in range(n):
                y_k += c[i]*b(d, t, i, delx)
            y.append(y_k)
        return y

    x = np.linspace(t[0], t[-1], 1000)

    y = bspline(c, t, x, d)

    plt.figure()
    plt.plot(x, y, color='green')

    ax = plt.axes()
    plt.xticks(t)
    ax.xaxis.grid() # vertical lines
    plt.axvline(t[d],  color='red')
    plt.axvline(t[n+1-d],  color='red')
    plt.xlabel('t')
    plt.ylabel('x')
    plt.savefig('spline_curve'+str(d)+'.pdf')
bspline_plot(3)
```
